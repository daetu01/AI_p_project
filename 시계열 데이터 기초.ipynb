```python
np.random.seed(1) #고정된 난수 출력을 위해 시드 설정

arr = np.random.randn(365) # 365일치 난수 생성
time_idx = pd.date_range('2021-1-1', periods=365, freq='D')

ts = pd.Series(arr, index = time_idx) #1년치 시계열 데이터 생성
print(ts)
```

```python
# down-sampling
# 다운샘플링 예제
ts.resample('M').mean() #일별 데이터를 월별로 변환
```

```python
# 업샘플링 예제 
arr = np.random.randn(10) #새로운 난수 10개 생성
time_idx = pd.date_range('2021-1-1', periods=10, freq='W')

ts = pd.Series(arr, index=time_idx)
print(ts)
```

```python
# forward filling 방식 
ts.resample('D').ffill().head(21) #ffill: 각 기간의 첫 일을 참고하여 결측값 보관
```

```python
# backward filling : 각 기간의 마지막 날을 참고하여 결측값 보관
ts.resample('D').bfill().head(21)
```

```python
# 필요 데이터 사용
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt # 그래프 상세 속성 설정 
```

```python
np.random.seed(1) # 난수 설정

ts_data = np.random.randn(365, 3) # (365 x 3) 행렬의 난수 데이터를 생성
```

```python
ts_idx = pd.date_range(start='2021-01-01', periods=365, freq='D') #365개의 시계열 인덱스 생성

ts_df = pd.DataFrame(ts_data, index = ts_idx) #시계열 데이터 프레임 
ts_df.head()
```

```python
ts_df
```

```python
ts_df.columns = ['X','Y','Z']
```

```python
ts_df
```

```python
ts_df.plot()
```

```python
ts_df.plot() #단순하게 데이터프레임의 메소드인 plot을 호출하면 됨 
plt.title('Time Series Plot of Random Numbers')
plt.show() #그래프 출력용 커맨드
```

```python
# 누적합 함수를 이용하여 알아보기 쉽게 각 시리즈를 변경
ts_df = ts_df.cumsum()
ts_df.plot()
plt.title('Time Series Plot of Random Numbers (Cumulative)')
plt.show()
```

```python
ts_df.plot(figsize=(10,3))
plt.title('Time Series Plot of Random Numbers')
plt.xlabel('Date') # x축 이름 지정
plt.ylabel('Unit') # y축 이름 지정
plt.show()
```
